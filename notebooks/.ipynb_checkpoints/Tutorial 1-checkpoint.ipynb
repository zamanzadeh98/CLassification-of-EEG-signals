## Fast fourier transform

```python
def fourierTransform(sig, hfreq):
    """
    This function apply fast fourier transform on
    signals

    parameters
    ------------
    sig: array
    signal array

    hfreq: int
    higher range of frequency res
    
    return
    ------------
    res: array
    fast fourier coefficent (real values and not complex values)
    """
    
    n = len(sig) # signal length

    # we should divide the fft result by n to get the right amplitude
    
    # we multiply it by 2 because the amplitudes is apportioned between
    # negative and positive frequencies
    # Also we take only half of the signal to take only the positive frequencies
    # why np.abs? only real values :)
    res = (2/n)*np.abs(scipy.fft.fft(subj[0][0])[0: n//2])

    # Since in EEG we don't need very high frequency values:
    res = res[:hfreq]
    
    return res
```

```python
res = fourierTransform(subj[0][0], hfreq=200)

fig, ax = plt.subplots(1,2, figsize = (13, 3))
ax[0].plot(subj[0][0], color="green")
ax[0].set_title("Time domain")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Amplitude")
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)

ax[1].plot(res, color='green')
ax[1].set_title("Frequency domain")
ax[1].set_xlabel("Frequency")
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
plt.show()
```

## Power sepctral analysis using Welch's method
Welch's method is fairly similar to Fourier transform. However, you have to convlove a short winodw along the signal and compute FFT for each. Then all you have to do is averaging!

**Remember**: It's good to have some overlap between windows. This way, you are less likely to lose information!

```python
fs = 500 #HZ :sampling rate (quite large!)
nperseg = fs*4
frqArr, psd = scipy.signal.welch(subj[0][0], fs=fs, nperseg=nperseg)

plt.figure(figsize=(10,3))
plt.plot(frqArr, psd, color="black")
plt.xlabel("Freq")
plt.ylabel("PSD (unit: V^2/Hz)")
plt.show()
```

## Instantaneous phase

Instantaneous phase refers to the phase of a signal at a specific point in time. In signal processing, particularly in the context of time-varying signals, the instantaneous phase represents the phase angle of a signal at each instant or time point. 

Instantaneous phase can be obtained via hilbert transfrom

**instantanouse phase Synchrony:**
Instantaneous phase synchrony refers to the degree to which the instantaneous phases of two signals exhibit consistent and correlated changes over time

```python
electrode1 = subj[0][0]
electrode2 = subj[0][1]

insPhase1 = np.angle(hilbert(electrode1), deg=False)
insPhase2 = np.angle(hilbert(electrode2), deg=False)
insPhaseSynch = 1 - np.sin(np.abs(insPhase1 - insPhase2)/2)

fig, ax =plt.subplots(3,1, figsize=(10,4))

ax[0].plot(electrode1, color="black", label="electrode 1", linewidth=0.5)
ax[0].plot(electrode2, color="red", label="electrode 2",  linewidth=0.5)
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].legend()


ax[1].plot(insPhase1, color="black", label="Instantaneous phase electrode 1", linewidth=0.5)
ax[1].plot(insPhase2, color="red", label="Instantaneous phase  electrode 2", linewidth=0.5)
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].legend()

ax[2].plot(insPhaseSynch, color="green", label="Instantaneous phase Synchrony", linewidth=1)
ax[2].spines['top'].set_visible(False)
ax[2].spines['right'].set_visible(False)
ax[2].legend()
```

## Spectogram

I love spectograms because they convert 1D signals into images, 2 dimensional signals :)

These pictures show frquency changes over time instead of a static representation of whole signal. Isn't that great? 

Spectogram leverage the power behind Short Fourier transform algorithm.

```python
f, t, Sxx = scipy.signal.spectrogram(electrode1, fs=500)

fig, ax = plt.subplots(2,1, figsize=(10,4))

ax[0].plot(electrode1, color="black", linewidth=0.6)
ax[0].set_xlabel("Sample")
ax[0].set_ylabel("Amplitude")

ax[1].pcolormesh(t, f, 10 * np.log10(Sxx), shading='gouraud')
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Frequency")
ax[1].set_ylim(0,100)
```

## Wavelet
The beatiful one!

The problem with Fourier transform is that it's only easily interpretable for stationary signals. In other words, although it has a high freqency resolution, it has low time resolution :). It's sad, isn't it?

The idea is that for lower frequencies we neglect time resolution and as we increase the frequency, we increase the time resolution and neglect frequency resolution!

```python
cA = electrode1.copy()

fig, ax = plt.subplots(15, 2, figsize=(20,20))

for i in range(15):
    # approxiamtion coeff, detail coeff
    cA, cD = pywt.dwt(cA, wavelet="db2")

    ax[i, 0].plot(cA, color="black", linewidth=0.5)
    
    ax[i, 1].plot(cD, color="green", linewidth=0.5)

    if i == 0:
        ax[i, 0].set_title("Approximation coefficient")
        ax[i, 1].set_title("Detail coefficient")

plt.tight_layout()
```
